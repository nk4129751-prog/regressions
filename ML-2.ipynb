```python
import numpy as np
x = np.array([2, 4, 6, 8 ])  # years of  experience
y = np.array([20, 40, 60, 80 ])   # salary in thousands
# Main formula --> y = mx+b
# formula for slope  --> m = mean(xy)-mean(x)*mean(Y)/mean(x**2)-(mean(x))2
# formula for intercept --> b = mean(Y)-m*mean(x)
# Sol:
multiplication = x*y
print("1.multiplication:" , multiplication)
meanofxy = np.mean(multiplication)
print("2.mean of xy : " , meanofxy)
meanx = np.mean(x)
print("3.mean of x : " ,meanx)
meany = np.mean(y)
print("4.mean of y : " , meany)
valuesofmean = meanx*meany
print("5.values of mean : " , valuesofmean)
newvaluesofnumerator = meanofxy - valuesofmean
print("6.new values of numerator : " ,newvaluesofnumerator )
meanofxsquare = np.mean(x**2)
print("7.meanofsquare : " , meanofxsquare)
squareofxmean = meanx**2
print("8.square of mean : " , squareofxmean)
newvaluesofdenominator = meanofxsquare - squareofxmean
print("9.new values of denominator : " , newvaluesofdenominator)

# put all the values in the formula of M 
m = newvaluesofnumerator / newvaluesofdenominator
print("10.Value of slope : " , m)

intercept = meany - m*meanx
print("11.value of intercept : " , intercept)
# put all the values of slope and intercept in main formula
predictedvalue = m*7 + intercept
print("12.Salary for 7 years experience : " , predictedvalue)
print("13.salary for 10 years experience : " , m*10 + intercept)
print("14.salary for 0 years experience : " , m*0 + intercept)

```
